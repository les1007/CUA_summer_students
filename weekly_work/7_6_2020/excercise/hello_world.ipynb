```python
x = [1,2,3,4]
for val in x:
    print(val)
```

```python
y = [5,6,7,8]
for val in y:
    print(val)
```

```python
import numpy as np
x_new = np.array(x)
print(x_new)
```

```python
y_new = np.array(y)
print(y_new)
```

```python
print(x_new)
print(y_new)
```

```python
import matplotlib.pyplot as plt
plt.plot(x_new,y_new)
plt.title('x vs y', fontsize =20)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.scatter(x_new,y_new)
plt.title('x vs y', fontsize =20)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
m,b = np.polyfit(x_new,y_new,1)
plt.scatter(x_new,y_new)
plt.plot(x_new, m*x_new + b)
plt.title('x vs y', fontsize =20)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
