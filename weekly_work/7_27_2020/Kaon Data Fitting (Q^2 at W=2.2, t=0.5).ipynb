```python
import matplotlib.pyplot as plt
import math
import numpy as np

pts = np.array([(1.90,2.35),(162.7,118.1),(11.9,9.0)])

x = pts[0,:]
y = pts[1,:]
yerror = pts[2,:]

x_min = pts[0,:]
y_min = (pts[1,0]+pts[2,0],pts[1,1]-pts[2,1])
y_min = np.array(y_min)
x_max = pts[0,:]
y_max = (pts[1,0]-pts[2,0],pts[1,1]+pts[2,1])
y_max = np.array(y_max)

m,b = np.polyfit(x,y,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix = np.corrcoef(x,y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x, m*x + b, label='Best Fit Line (y = -99.1x + 351.0)($R^2$=1.0)')
plt.plot(x_min, m1*x_min + b1, linestyle='--', dashes=(3,5), label='Minimum Slope (y = -145.6 + 451.2)')
plt.plot(x_max, m2*x_max + b2, linestyle='--', dashes=(3,5), label='Maximum Slope (y = -52.7x + 250.9)')
plt.legend()
plt.errorbar(x,y,yerror,fmt=' ', color='blue')
plt.title('$Q^2$ Dependence at W=2.2, t=0.5', fontsize =20)
plt.xlabel('$Q^2$ ($GeV^2$)', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```
