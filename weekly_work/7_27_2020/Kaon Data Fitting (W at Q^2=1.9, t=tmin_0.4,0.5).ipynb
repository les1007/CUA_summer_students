```python
import numpy as np
import matplotlib.pyplot as plt

pts = np.array([(1.94,2.00,2.14),(170.0,162.7,152.1),(12.3,11.9,12.8)])

x = pts[0,:]
y = pts[1,:]
y_error = pts[2,:]

m,b = np.polyfit(x,y,1)
print(m,b)

correlation_matrix = np.corrcoef(x,y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.plot(x, m*x + b, label='Best Fit Line (y = -87.1x + 338.0)($R^2$=0.985)')
plt.legend()
plt.errorbar(x,y,y_error,fmt=' ')
plt.title('W Dependence at $Q^2$=1.9, t=tmin~0.4/0.5', fontsize =20)
plt.xlabel('W (GeV)', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

```python
x_2 = 1/(pts[0,:]**2)
y_2 = pts[1,:]
y_2error = pts[2,:]

x_min = (1/pts[0,0]**2,1/pts[0,2]**2)
x_min = np.array(x_min)
y_min = (pts[1,0]-pts[2,0],pts[1,2]+pts[2,2])
y_min = np.array(y_min)
x_max = (1/pts[0,0]**2,1/pts[0,2]**2)
x_max = np.array(x_max)
y_max = (pts[1,0]+pts[2,0],pts[1,2]-pts[2,2])
y_max = np.array(y_max)

m,b = np.polyfit(x_2,y_2,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
print('\n')
print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix = np.corrcoef(x_2,y_2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_2,y_2)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_2, m*x_2 + b, label='Best Fit Line (y = 371.9x + 70.6)($R^2$=0.993)')
plt.plot(x_min, m1*x_min + b1, dashes=(3,5), label='Min Slope (y = -152.1x + 198.1)')
plt.plot(x_max, m2*x_max + b2, dashes=(3,5), label='Max Slope (y = 908.3x - 59.0)')
plt.legend()
plt.errorbar(x_2,y_2,y_error,fmt=' ', color='blue')
plt.title(r'Kaon $\frac{1}{W^2}$ Dependence at $Q^2$=1.9, t=tmin~0.4/0.5', fontsize =20)
plt.xlabel(r'$\frac{1}{W^2}$ ($\frac{1}{(GeV)^2}$)', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```
