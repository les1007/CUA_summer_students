```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.172,0.252,0.149),(376.6,940,1340),(12.0,160,100)])

x = pts[0,:]
y = pts[1,:]
yerror = pts[2,:]

m,b = np.polyfit(x,y,1)
print(m,b)

correlation_matrix = np.corrcoef(x,y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.plot(x, m*x + b, label='Best Fit Line (y = -1043x + 1085)($R^2$=0.014)')
plt.legend()
plt.errorbar(x,y,yerror,fmt=' ')
plt.title('Fig1. x Dependence at $Q^2$=1.6, W=2.2, t=0.2, x=0.1-0.2', fontsize =20)
plt.xlabel('x', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

Figure 1 shows the plot of the raw kaon data, x at $Q^2$=1.6, W=2.2, t=0.2, and x=0.1-0.2. The line of best fit shows a very low $R^2$ value of 0.014.

```python
x_2 = pts[0,:]**2
y_2 = np.log(pts[1,:])
y_2error = y_2*(pts[2,:]/pts[1,:])
x_min = (x_2[2],x_2[1])
x_min = np.array(x_min)
y_min = (y_2[2]+y_2error[2],y_2[1]-y_2error[1])
y_min = np.array(y_min)
x_max = (x_2[2],x_2[1])
x_max = np.array(x_max)
y_max = (y_2[2]-y_2error[2],y_2[1]+y_2error[1])
y_max = np.array(y_max)

m,b = np.polyfit(x_2,y_2,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix2 = np.corrcoef(x_2,y_2)
correlation_xy2 = correlation_matrix2[0,1]
r_squared2 = correlation_xy2**2
print(r_squared2)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_2,y_2)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_2, m*x_2 + b, label='Best Fit Line (y = 2.4x + 6.6)($R^2$=0.007)')
plt.plot(x_min, m1*x_min + b1, dashes=(3,5), label='Minimum Slope (y = -49.8x + 8.8)')
plt.plot(x_max, m2*x_max + b2, dashes=(3,5), label='Maximum Slope (y = 32.6x + 5.9)')
plt.legend()
plt.errorbar(x_2,y_2,y_2error,fmt=' ', color='blue')
plt.title('Fig2. $x^2$ at $Q^2$=1.6, W=2.2, t=0.2, x=0.1-0.2', fontsize =20)
plt.xlabel('$x^2$', fontsize =15)
plt.ylabel(r'ln$\sigma$ (ln$\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

Figure 2 shows $x^2$ plotted against ln$\sigma$. This fit also has a very low $R^2$ value of 0.007 due to the 2nd point. The line of best fit is y = 2.4x + 6.6. It has a very large slope uncertainty of 1675%. With more data collection with smaller uncertainties, a better fit with a smaller slope uncertainty may be found.
