```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.30,0.41,0.414),(329.7,293.9,152.1),(10.8,10.4,12.8)])

x = pts[0,:]
y = pts[1,:]
yerror = pts[2,:]

m,b = np.polyfit(x,y,1)
print(m,b)

correlation_matrix = np.corrcoef(x,y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.plot(x, m*x + b, label='Best Fit Line (y = -985.6x + 627.9)($R^2$=0.461)')
plt.legend()
plt.errorbar(x,y,yerror,fmt=' ')
plt.title('Fig1. t Dependence at $Q^2$=1.6, W=2.2, t=0.3-0.4, x=0.3-0.5', fontsize =20)
plt.xlabel(r't ($\frac{GeV}{c}$)', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

```python
x_2 = pts[0,:]**6
y_2 = pts[1,:]
y_2error = y_2*(pts[2,:]/pts[1,:])

m,b = np.polyfit(x_2,y_2,1)
print(m,b)

correlation_matrix = np.corrcoef(x_2,y_2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_2,y_2)
plt.plot(x_2, m*x_2 + b, label='Best Fit Line (y = -27279x + 354)($R^2$=0.489)')
plt.errorbar(x_2,y_2,y_2error,fmt=' ')
plt.legend()
plt.title('Fig2. $t^6$ Dependence at $Q^2$=1.6, W=2.2, t=0.3-0.4, x=0.3-0.5', fontsize =20)
plt.xlabel(r'$t^6$ (($\frac{GeV}{c})^6)$', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

Figure 1 shows the raw kaon data plot, t at $Q^2$=1.6, W=2.2, t=0.3-0.4, and x=0.3-0.5. This plot has a best fit line of y = -985.6x + 627.9 with an $R^2$ value of 0.461, which is not a very good $R^2$ value to have. When the data is plotted to the fit of $t^6$, as shown in figure 2, the $R^2$ value increased to 0.489. This is not the best fit because it has a very low $R^2$ value but this is the best fit that can be found with the data points provided. More data collection will make the best fit more evident.
