```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(1.90,2.00,2.35),(170.7,184.5,135.1),(15.2,8.0,11.6)])

x = pts[0,:]
y = pts[1,:]
yerror = pts[2,:]

m,b = np.polyfit(x,y,1)
print(m,b)

correlation_matrix = np.corrcoef(x,y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.plot(x, m*x + b, label='Best Fit Line (y = -95.3x + 362.0)($R^2$=0.781)')
plt.legend()
plt.errorbar(x,y,yerror,fmt=' ')
plt.title('Fig1. $Q^2$ Dependence at W=2.2, t=0.7', fontsize =20)
plt.xlabel('$Q^2$ ($GeV^2$)', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

```python
x_2 = pts[0,:]**2
y_2 = np.log(pts[1,:])
y_2error = y_2*(pts[2,:]/pts[1,:])
x_min = (1.90**2,2.35**2)
x_min = np.array(x_min)
y_min = (y_2[0]+y_2error[0],y_2[2]-y_2error[2])
y_min = np.array(y_min)
y_min = abs(y_min)
x_max = (1.90**2,2.35**2)
x_max = np.array(x_max)
y_max = (y_2[0]-y_2error[0],y_2[2]+y_2error[2])
y_max = np.array(y_max)
y_max = abs(y_max)

m,b = np.polyfit(x_2,y_2,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix = np.corrcoef(x_2,y_2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_2,y_2)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_2, m*x_2 + b, label='Best Fit Line (y = -0.1x + 5.7)($R^2$=0.821)')
plt.plot(x_min, m1*x_min + b1, dashes=(3,5), label='Minimum Slope (y = -0.6x + 7.7)')
plt.plot(x_max, m2*x_max + b2, dashes=(3,5), label='Maximum Slope (y = 0.3x + 3.5)')
plt.legend()
plt.errorbar(x_2,y_2,y_2error,fmt=' ', color='blue')
plt.title('Fig2. $Q^4$ Dependence at W=2.2, t=0.7', fontsize =20)
plt.xlabel('$Q^4$ ($(GeV^2)^2$)', fontsize =15)
plt.ylabel(r'ln$\sigma$ (ln$\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

```python
x_3 = 1/(pts[0,:]**3)
y_3 = pts[1,:]
y_3error = pts[2,:]
x_min = (x_3[2],x_3[0])
x_min = np.array(x_min)
y_min = (y_3[2]+y_3error[2],y_3[0]-y_3error[0])
y_min = np.array(y_min)
x_max = (x_3[2],x_3[0])
x_max = np.array(x_max)
y_max = (y_3[2]-y_3error[2],y_3[0]+y_3error[0])
y_max = np.array(y_max)

m,b = np.polyfit(x_3,y_3,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix = np.corrcoef(x_3,y_3)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_3,y_3)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_3, m*x_3 + b, label='Best Fit Line (y = 607.4x + 93.0)($R^2$=0.706)')
plt.plot(x_min, m1*x_min + b1, dashes=(3,5), label='Minimum Slope (y = 128.0x + 136.8)')
plt.plot(x_max, m2*x_max + b2, dashes=(3,5), label='Maximum Slope (y = 907.8x + 53.6)')
plt.legend()
plt.errorbar(x_3,y_3,y_3error,fmt=' ', color='blue')
plt.title(r'$\frac{1}{Q^6}$ Dependence at W=2.2, t=0.7', fontsize =20)
plt.xlabel(r'$\frac{1}{Q^6}$ ($\frac{1}{(GeV^2)^3}$)', fontsize = 15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

Figure 1 shows the plot of the raw data for this set of kaon data, $Q^2$ at W=2.2 and t=0.7. From comparing figure 2 and 3, the fit of $Q^4$ vs ln$\sigma$ shows to be a better fit than the expected fit of $\frac{1}{Q^6}$. Figure 2 shows that when $Q^4$ is fitted aginst ln$\sigma$, the equation of the line of best fit is y = -0.1x + 5.7 and the $R^2$ value is 0.821. In figure 3, when $\frac{1}{Q^6}$ is plotted against $\sigma$, the $R^2$ value is 0.706. Thus, it seems like $Q^4$ vs ln$\sigma$ is a better fit because it has a larger $R^2$ value. However, figure 2 shows a slope uncertainty of 350% while figure 3 shows a slope uncertainty of 64%. Therefore, it is difficult to tell which fit is a better fit for this data set. More data collection with smaller errorbars could yield better results.
