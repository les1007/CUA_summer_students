```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.379,0.405,0.397,0.407,0.436),(162.7,118.1,170.0,170.7,135.1),(11.9,9.0,12.3,15.2,11.6)])

x = pts[0,:]
y = pts[1,:]
yerror = pts[2,:]

m,b = np.polyfit(x,y,1)
print(m,b)

correlation_matrix = np.corrcoef(x,y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.plot(x, m*x + b, label='Best Fit Line (y = -533.4x + 367.2)($R^2$=0.218)')
plt.legend()
plt.errorbar(x,y,yerror,fmt=' ')
plt.title('Fig1. x Dependence at $Q^2$=1.6, W=2.2, t=0.5-0.6, x=0.3-0.5', fontsize =20)
plt.xlabel('x')
plt.ylabel(r'$\sigma$ $\frac{nbar}{c}$')
plt.show()
```

Plot of the raw data. Graph shows the line of best fit at y = -533.4x + 367.2 with an $R^2$ value of 0.218.

```python
x_2 = pts[0,:]**2
y_2 = pts[1,:]
y_2error = pts[2,:]
x_min = (x_2[0],x_2[4])
x_min = np.array(x_min)
y_min = (y_2[0]+y_2error[0],y_2[4]-y_2error[4])
y_min = np.array(y_min)
x_max = (x_2[0],x_2[4])
x_max = np.array(x_max)
y_max = (y_2[0]-y_2error[0],y_2[4]+y_2error[4])
y_max = np.array(y_max)

m,b = np.polyfit(x_2,y_2,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix = np.corrcoef(x_2,y_2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_2,y_2)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_2, m*x_2 + b, label='Best Fit LIne (y = -652.6x + 258.5)($R^2$=0.218)')
plt.plot(x_min, m1*x_min + b1, dashes=(3,5), label='Minimum Slope (y = -1100.0x + 332.6)')
plt.plot(x_max, m2*x_max + b2, dashes=(3,5), label='Maximum Slope (y = -88.3x + 163.5)')
plt.legend()
plt.errorbar(x_2,y_2,y_2error, fmt=' ', color='blue')
plt.title('Fig2. $x^2$ Dependence at $Q^2$=1.6, W=2.2, t=0.5-0.6, x=0.3-0.5',fontsize =20)
plt.xlabel('$x^2$', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

Figure 2 shows a plot of $x^2$ vs $\sigma$. The best fit line is y = -652.6x + 258.5 and it has an $R^2$ value of 0.218. The poor $R^2$ value is due to the third point, which has a significantly smaller $\sigma$ value than the other data points. However, there is no physical explaination that can support that this point is an outlier so this point can not be taken out of consideration. More research is needed to explain the third data point; more data collection may be helpful.
