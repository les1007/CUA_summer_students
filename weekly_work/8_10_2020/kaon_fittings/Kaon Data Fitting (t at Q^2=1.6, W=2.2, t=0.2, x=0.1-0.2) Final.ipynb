```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.19,0.21,0.22,0.30,0.41,0.414),(940,1340,376.6,329.7,293.9,152.1),(160,100,12.0,10.8,10.4,12.8)])

x = pts[0,:]
y = pts[1,:]
yerror = pts[2,:]

m,b = np.polyfit(x,y,1)
print(m,b)

correlation_matrix = np.corrcoef(x,y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.plot(x,m*x + b, label='Best Fit Line (y = -3369x + 1551)($R^2$=0.540)')
plt.legend()
plt.errorbar(x,y,yerror,fmt=' ')
plt.title('Fig1. t at $Q^2$=1.6, W=2.2, t=0.2, x=0.1-0.2', fontsize =20)
plt.xlabel(r't (($\frac{GeV}{c})^2$)', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

```python
x_2 = 1/(pts[0,:]**2)
y_2 = np.log(pts[1,:])
y_2error = y_2*(pts[2,:]/pts[1,:])
x_min2 = (x_2[5],x_2[0])
x_min2 = np.array(x_min2)
y_min2 = (y_2[5]+y_2error[5],y_2[0]-y_2error[0])
y_min2 = np.array(y_min2)
x_max2 = (x_2[5],x_2[0])
x_max2 = np.array(x_max2)
y_max2 = (y_2[5]-y_2error[5],y_2[0]+y_2error[0])
y_max2 = np.array(y_max2)

m,b = np.polyfit(x_2,y_2,1)
m1,b1 = np.polyfit(x_min2,y_min2,1)
m2,b2 = np.polyfit(x_max2,y_max2,1)
print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix = np.corrcoef(x_2,y_2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_2,y_2)
plt.scatter(x_min2,y_min2)
plt.scatter(x_max2,y_max2)
plt.plot(x_2, m*x_2 + b, label='Best Fit Line (y = 0.07x + 4.92)($R^2$=0.732)')
plt.plot(x_min2, m1*x_min2 + b1, dashes=(3,5), label='Minimum Slope (y = 0.01x + 5.39)')
plt.plot(x_max2, m2*x_max2 + b2, dashes=(3,5), label='Maximum Slope (y = 0.16x + 3.69)')
plt.legend()
plt.errorbar(x_2,y_2,y_2error,fmt=' ', color='blue')
plt.title(r'Fig2. $\frac{1}{t^2}$ Dependence at $Q^2$=1.6, W=2.2, t=0.2, x=0.1-0.2', fontsize =20)
plt.xlabel(r'$\frac{1}{t^2}$ (($\frac{c}{GeV})^4)$', fontsize =15)
plt.ylabel(r'ln$\sigma$ (ln$\frac{nbar}{c}$)', fontsize =15)
plt.ylim(4.3,8.3)
plt.show()
```

Figure 1 shows the plot of the raw kaon data set, t at $Q^2$=1.6, t=0.2, and x=0.1-0.2. This data is fitted on a $\frac{1}{t^2}$ vs ln$\sigma$ plots in figure 2. The equation of the line of best fit is y = 0.07x + 4.92 with an $R^2$ value of 0.732. Figures 3 through 5 show the fit $\frac{1}{t^2}$ vs ln$\sigma$ like in figure 2 but with different data point excluded in each figure. The purpose of this excercise was to investigate the effect of errorbars on the fit of the data. when a data point with the largest errorbar is excluded as shown in figure 3, the $R^2$ value did not change dramatically. However, when the data point with a smaller errorbar is taken out as displayed in figures 4 and 5, there is a dramatic change in the $R^2$ value, either increasing or decreasing. From this investigation, one can conclude that the goodness of the fit, indicated by the $R^2$ value, is driven by the small errorbars. Thus, data collected with smaller errors can yield more accurate fits or results.
