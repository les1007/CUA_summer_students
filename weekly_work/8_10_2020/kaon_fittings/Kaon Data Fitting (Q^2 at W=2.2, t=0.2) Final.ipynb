```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.52,0.70,1.35),(376.6,1340,940),(10,100,200),(12.0,137,173)])

x = pts[0,:]
y = pts[1,:]
yerror = pts[2,:]

m,b = np.polyfit(x,y,1)
print(m,b)

correlation_matrix = np.corrcoef(x,y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.plot(x, m*x + b, label='Best Fit Line (y = 333x + 600)($R^2$=0.090)')
plt.legend()
plt.errorbar(x,y,yerror,fmt=' ')
plt.title('Fig1. $Q^2$ Dependence at W=2.2, t=0.2', fontsize =20)
plt.xlabel('$Q^2$ ($GeV^2$)', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

```python
x_2 = pts[0,:]**2
y_2 = np.log(pts[1,:])
y_2error = y_2*(pts[2,:]/pts[1,:])

m,b = np.polyfit(x_2,y_2,1)
print(m,b)

correlation_matrix = np.corrcoef(x_2,y_2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_2,y_2)
plt.plot(x_2, m*x_2 + b, label='Best Fit Line (y = 0.3x + 6.4)($R^2$=0.138)')
plt.legend()
plt.errorbar(x_2,y_2,y_2error,fmt=' ')
plt.title('Fig2. $Q^4$ Dependence at W=2.2, t=0.2', fontsize =20)
plt.xlabel('$Q^4$ ($(GeV^2)^2$)', fontsize =15)
plt.ylabel(r'ln$\sigma$ (ln$\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

```python
x_3 = 1/(pts[0,:]**3)
y_3 = pts[1,:]
y_3error = pts[2,:]

m,b = np.polyfit(x_3,y_3,1)
print(m,b)

correlation_matrix = np.corrcoef(x_3,y_3)
correlaiton_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_3,y_3)
plt.plot(x_3, x*x_3 + b, label='Best Fit Line (y = -99.0x + 1229.8)($R^2$=0.138)')
plt.legend()
plt.errorbar(x_3,y_3,y_3error,fmt=' ')
plt.title(r'Fig3. $\frac{1}{Q^6}$ Dependence at W=2.2, t=0.2', fontsize =20)
plt.xlabel(r'$\frac{1}{Q^6}$ ($\frac{1}{(GeV^2)^3}$)', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

For this kaon set, $Q^2$ at W=2.2, t=0.2, two fits were tried as shown in figure 2 and figure 3. Figure 2 shows $Q^4$ against ln$\sigma$; it has a best fit line equation of y = 0.3x + 6.4 and an $R^2$ value of 0.138. The errorbars on the data are very large so this indicates the fit might not be the best. In figure 3, the data is plotted to the expcted fit, $\frac{1}{Q^6}$. It is interesting that the two fits in figure 2 and 3 shows the same $R^2$ values. However, the errorbars in figure 3 are much smaller. Thus, with more data points, $\frac{1}{Q^6}$ might appear to be a better fit than $Q^4$.
