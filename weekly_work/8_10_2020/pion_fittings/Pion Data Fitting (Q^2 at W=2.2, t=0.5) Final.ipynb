```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(3.766,3.44),(0.721,0.0183),(0.060,0.007)])
y_min = (0.661,0.0253)
y_max = (0.781,0.0113)

x_1 = pts[0,:]
y_1 = pts[1,:]
y_1error = pts[2,:]
x_min = pts[0,:]
y_min = np.array(y_min)
x_max = pts[0,:]
y_max = np.array(y_max)

m,b = np.polyfit(x_1,y_1,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix = np.corrcoef(x_1,y_1)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_1,y_1)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_1, m*x_1 + b, label='Best Fit Line (y = 2.156x - 7.397)($R^2$=1)')
plt.plot(x_min, m1*x_min + b1, dashes=(3,5), label='Min Slope (y = 1.950x - 6.683)')
plt.plot(x_max, m2*x_max + b2, dashes=(3,5), label='Max Slope (y = 2.361x - 8.111)')
plt.legend()
plt.errorbar(x_1,y_1,y_1error, fmt=' ', color='blue')
plt.title('Fig1. $Q^2$ Dependence at W=2.2, t=0.5', fontsize =20)
plt.xlabel('$Q^2$ ($GeV^2$)', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

This data set, $Q^2$ at W=2.2 and t=0.5, only has two points so it is hard to make any conclusions about it. With more data points at W=2.2 and t=0.5, the fit $\frac{1}{Q^6}$ can be tried. Currently, Fig1 shows a linear fit of the two data points. The equation of the fit is y = 2.156x - 7.39 with a slope uncertainty of 9.5%.
