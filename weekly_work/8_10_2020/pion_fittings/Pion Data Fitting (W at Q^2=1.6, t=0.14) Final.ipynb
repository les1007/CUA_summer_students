```python
import numpy as np
import matplotlib.pyplot as plt

x = [2.001,2.397,2.213]
x_1 = np.array(x)
y = [7.152,2.807,4.292]
y_1 = np.array(y)
y_1error = (0.356,0.124,0.473)
y_1error = np.array(y_1error)

x_min = [2.001,2.397]
y_min = [6.796,2.937]
x_min = np.array(x_min)
y_min = np.array(y_min)
x_max = [2.001,2.397]
y_max = [7.508,2.673]
x_max = np.array(x_max)
y_max = np.array(y_max)

plt.figure(figsize=(8,5))
plt.plot(x_1,y_1,'o')
plt.errorbar(x_1,y_1,y_1error, fmt=' ')
plt.title('W Dependence at $Q^2=1.6$, t=0.14', fontsize =20)
plt.xlabel('W (GeV)', fontsize =15)
plt.ylabel('$\sigma$ (nbar/c)', fontsize =15)
plt.show()
```

```python
x_3 = [1/(2.001**2),1/(2.397**2),1/(2.213**2)]
x_3 = np.array(x_3)
y_3 = y_1
x_min = [1/(2.397**2),1/(2.001*2.001)]
x_min = np.array(x_min)
y_min = [2.931,6.796]
y_min = np.array(y_min)
x_max = [1/(2.397*2.397),1/(2.001*2.001)]
x_max = np.array(x_max)
y_max = [2.683,7.508]
y_max = np.array(y_max)

m,b = np.polyfit(x_3,y_3,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
print(m,b)
print(m1,b1)
print(m2,b2)
correlation_matrix = np.corrcoef(x_3,y_3)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_3,y_3)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_3, m*x_3 + b, label='Best Fit Line (y = 57.828x - 7.354)($R^2=0.996$)')
plt.plot(x_min, m1*x_min + b1, linestyle='--', dashes=(3,5), label='Min Slope (y = 51.053x - 5.955)')
plt.plot(x_max, m2*x_max + b2, linestyle='--', dashes=(3,5), label='Max Slope (y = 63.735x - 8.410)')
plt.legend()
plt.errorbar(x_3,y_3,y_1error, fmt=" ", color='blue')
plt.title(r"Pion $\frac{1}{W^2}$ Dependence at $Q^2=1.6$, t=0.14",fontsize =20)
plt.xlabel(r"$\frac{1}{W^2}$ ($\frac{1}{(GeV)^2}$)", fontsize =15)
plt.ylabel(r"$\sigma$ ($\frac{nbar}{c}$)", fontsize=15)
plt.show()
```

In figure 1, the center of mass energy (W) is plotted against sigma, representing the traseverse area that the electron must hit in order for the pion to bounce off in a certain angle. When plotted, they seem to show a inverse relationship. The fit appears to be the best when $\frac{1}{W^2}$ is plotted against $\sigma$, giving an R-squared value of 0.996. The equation of this fit is y = 57.828x - 7.354 and the slope uncertainy of 11.0%. Another indicator that supports this fit is that the line of best fit falls within the errorbars of all data points. Thus, it can be concluded that W and $\sigma$ have a inverse-square relationship.
