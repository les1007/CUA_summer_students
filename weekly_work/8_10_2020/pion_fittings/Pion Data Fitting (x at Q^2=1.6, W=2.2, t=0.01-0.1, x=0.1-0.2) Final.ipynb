```python
import matplotlib.pyplot as plt
import numpy as np

pts=np.array([(0.090,0.147,0.152,0.164,0.174,0.184),(34.1,27.49,11.39,25.3,22.7,17.5),(4.3,2.5,1.6,2.2,2.0,1.86)])

x= pts[0,:]
y= pts[1,:]
yerror= pts[2,:]

m,b = np.polyfit(x,y,1)
print(m,b)

correlation_matrix = np.corrcoef(x,y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.plot(x, m*x + b, label='Best Fit Line (y = -156.8x + 46.9)($R^2$=0.433)')
plt.legend()
plt.errorbar(x,y,yerror, fmt=' ')
plt.title('Fig1. x Dependence at $Q^2$=1.6, W=2.2, t=0.01-0.1, x=0.1-0.2', fontsize=20)
plt.xlabel('x', fontsize=15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize=15)
plt.show()
```

```python
x_2 = pts[0,:]
y_2 = np.log(pts[1,:])
y_2error = y_2*(pts[2,:]/pts[1,:])
x_min = (pts[0,0],pts[0,5])
x_min = np.array(x_min)
y_min = (y_2[0]+y_2error[0],y_2[5]-y_2error[5])
y_min = np.array(y_min)
x_max = (pts[0,0],pts[0,5])
x_max = np.array(x_max)
y_max = (y_2[0]-y_2error[0],y_2[5]+y_2error[5])
y_max = np.array(y_max)

m,b = np.polyfit(x_2,y_2,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix = np.corrcoef(x_2,y_2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_2,y_2)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_2, m*x_2 + b, label='Best Fit Line (y = -6.0x + 4.0)($R^2$=0.268)')
plt.plot(x_min, m1*x_min + b1, linestyle='--', dashes=(3,5), label='Minimum Slope (y = -15.1x + 5.3)')
plt.plot(x_max, m2*x_max + b2, linestyle='--', dashes=(3,5), label='Maximum Slope (y = 0.9x + 3.0)')
plt.legend()
plt.errorbar(x_2,y_2,y_2error,fmt=' ', color='blue')
plt.title('Fig2. x at $Q^2$=1.6, W=2.2, t=0.01-0.1, x=0.1-0.2', fontsize =20)
plt.xlabel('x', fontsize =15)
plt.ylabel(r'ln$\sigma$ (ln$\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

This data set, x at $Q^2$=1.6, W=2.2, t=0.01-0.1, and x=0.1-0.2, was particularly difficult to find a line of best fit for because of the large error bars and the one data point that doesn't seems to follow the trend that the rest of the data points follow. However, it would be unreasonable to omit this point because the data table shows no indication that it is an outlier. Thus, it can not be explained with physics why this point would be an outlier so the point is displayed in both figures 1 and 2. Figure 2 shows a plot of x vs ln$\sigma$ with an $R^2$ value of 0.268. This is a very low $R^2$ value but the best fit line fits nicely within the errorbars of 5 of the 6 data points. Just looking at the $R^2$ value, it can be said that the fit of x vs $\sigma$ as displayed in figure 1 is a better fit. Figure 1 shows that the $R^2$ value for this fit is 0.433.
