```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.70,0.646,0.781,0.945,1.416,1.487),(10.4,17.51,15.79,13.63,6.44,4.179),(1.8,1.86,1.93,1.29,0.68,0.169)])

x_1 = pts[0,:]
y_1 = pts[1,:]
y_1error = pts[2,:]

m,b = np.polyfit(x_1,y_1,1)
print(m,b)

correlation_matrix = np.corrcoef(x_1,y_1)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_1,y_1)
plt.plot(x_1, m*x_1 + b, label='Best Fit Line (y = -12.6x + 23.9)($R^2$ = 0.772)')
plt.legend()
plt.errorbar(x_1,y_1,y_1error,fmt=' ')
plt.title('Fig1. $Q^2$ Dependence at W=2.2, t=0.08', fontsize=20)
plt.xlabel('$Q^2$ ($GeV^2$)', fontsize=15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize=15)
plt.show()
```

```python
x_2 = pts[0,:]**3
y_2 = np.log(pts[1,:])
y_2error = y_2*(pts[2,:]/pts[1,:])
x_min = (x_2[1],x_2[5])
x_min = np.array(x_min)
y_min = (y_2[1]+y_2error[1],y_2[5]-y_2error[5])
y_min = np.array(y_min)
x_max = (x_2[1],x_2[5])
x_max = np.array(x_max)
y_max = (y_2[1]-y_2error[1],y_2[5]+y_2error[5])
y_max = np.array(y_max)

m,b = np.polyfit(x_2,y_2,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix = np.corrcoef(x_2,y_2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_2,y_2)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_2, m*x_2 + b, label='Best Fit Line (y = -0.39x + 2.83)($R^2$=0.871)')
plt.plot(x_min, m1*x_min + b1, dashes=(3,5), label='Minimum Slope (y = -0.59x + 3.33)')
plt.plot(x_max, m2*x_max + b2, dashes=(3,5), label='Maximum Slope (y = -0.35x + 2.65)')
plt.legend()
plt.errorbar(x_2,y_2,y_2error, fmt=' ', color='blue')
plt.title('Fig2. $Q^6$ Dependence at W=2.2, t=0.08', fontsize=20)
plt.xlabel('$Q^6$ ($(GeV^2)^3$)', fontsize=15)
plt.ylabel(r'ln$\sigma$ (ln$\frac{nbar}{c}$)', fontsize=15)
plt.show()
```

```python
x_3 = 1/(pts[0,:]**3)
y_3 = pts[1,:]
y_3error = pts[2,:]

m,b = np.polyfit(x_3,y_3,1)
print(m,b)

correlation_matrix = np.corrcoef(x_3,y_3)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_3,y_3)
plt.plot(x_3, m*x_3 + b, label='Best Fit Line (y = 2.94x  6.15)($R^2$=0.600)')
plt.legend()
plt.errorbar(x_3,y_3,y_3error,fmt=' ')
plt.title(r'Fig3. $\frac{1}{Q^6}$ Dependence at W=2.2, t=0.08', fontsize =20)
plt.xlabel(r'$\frac{1}{Q^6}$ ($\frac{1}{(GeV^2)^3}$)', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

Figure 1, the raw data plot of $Q^2$ at W=2.2 and t=0.08, shows a linear fit at the equation y = -12.6x + 23.9 with an $R^2$ value of 0.772. However, a better fit can be observed when the data is fitted to $Q^6$ against ln$\sigma$. The equation of this fit is y = -0.39x + 2.83 with an increased $R^2$ value of 0.871 as shown in figure 2. This fit is better than the expected fit of $\frac{1}{Q^6}$, which had a best fit line equation of y = 2.94x + 6.15 with an $R^2$ value of 0.600. However, there is not enough evidence to conclude that $Q^6$ vs ln$\sigma$ is a better fit; more data collection can help determine whether it is the best fit or not.
