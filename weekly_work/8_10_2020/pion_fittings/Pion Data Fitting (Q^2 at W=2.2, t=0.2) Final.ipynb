```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.70,1.667,2.279,1.658,1.610),(4.79,3.91,2.34,2.521,5.443),(0.98,0.44,0.21,0.110,0.468)])

x_1 = pts[0,:]
y_1 = pts[1,:]
y_1error = pts[2,:]

m,b = np.polyfit(x_1,y_1,1)
print(m,b)

correlation_matrix = np.corrcoef(x_1,y_1)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_1,y_1)
plt.plot(x_1, m*x_1 + b, label='Best Fit Line (y = -1.51x + 6.20)($R^2$=0.392)')
plt.errorbar(x_1,y_1,y_1error, fmt=' ')
plt.legend()
plt.title('Fig1. $Q^2$ Dependence at W=2.2, t=0.2', fontsize=20)
plt.xlabel('$Q^2$ ($GeV^2$)', fontsize=15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize=15)
plt.show()
```

```python
x_2 = pts[0,:]**3
y_2 = np.log(pts[1,:])
y_2error = y_2*(pts[2,:]/pts[1,:])
x_max = (0.70**3,2.279**3)
x_max = np.array(x_max)
y_max = (y_2[0]-y_2error[0],y_2[2]+y_2error[2])
y_max = np.array(y_max)
x_min = (0.70**3,2.279**3)
x_min = np.array(x_min)
y_min = (y_2[0]+y_2error[0],y_2[2]-y_2error[2])
y_min = np.array(y_min)

m,b = np.polyfit(x_2,y_2,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix = np.corrcoef(x_2,y_2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_2,y_2)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_2, m*x_2 + b, label='Best Fit Line (y = -0.06x + 1.61)($R^2$=0.508)')
plt.plot(x_min, m1*x_min + b1, linestyle='--', dashes=(3,5), label='Min Slope (y = -0.10x + 1.92)')
plt.plot(x_max, m2*x_max + b2, linestyle='--', dashes=(3,5), label='Max Slope (y = -0.03x + 1.26)')
plt.errorbar(x_2,y_2,y_2error, fmt=' ', color='blue')
plt.legend()
plt.title('Fig2. $Q^6$ Dependence at W=2.2, t=0.2', fontsize=20)
plt.xlabel('$Q^6$ (($GeV^2)^3$)', fontsize=15)
plt.ylabel(r'ln$\sigma$ (ln$\frac{nbar}{c}$)', fontsize=15)
plt.show()
```

```python
x_3 = 1/(pts[0,:]**3)
y_3 = pts[1,:]
y_3error = pts[2,:]
m,b = np.polyfit(x_3,y_3,1)
print(m,b)

correlation_matrix = np.corrcoef(x_3,y_3)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_3,y_3)
plt.plot(x_3, m*x_3 + b, label='Best Fit Line (y = 0.49x + 3.44)($R^2$=0.189)')
plt.legend()
plt.errorbar(x_3,y_3,y_3error,fmt=' ')
plt.title(r'Fig3. $\frac{1}{Q^6}$ Dependence at W=2.2, t=0.2', fontsize =20)
plt.xlabel(r'$\frac{1}{Q^6}$ ($\frac{1}{(GeV^2)^3}$)', fontsize =15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize =15)
plt.show()
```

The expected fit for $Q^2$-dependence is $\frac{1}{Q^6}$. However, this particular pion data set, obtained at W=2.2 and t=0.2, has the best fit when $Q^6$ is plotted agianst ln$\sigma$ as shown in figure 2. Figure 2 shows that the best fit line equation of this fit is y = -0.06x + 1.61 with an $R^2$ value 0.508. Even though this is the best fit, the $R^2$ value is relatively low because of the 2nd and 3rd points that don't fall on the best fit line. However, there is no evidence that these are outliers. Thus, they are included in the analysis of this data set. When the data set is fitted to the expected fit, $\frac{1}{Q^6}$, as shown in figure 3, the $R^2$ value is 0.189. This supports that $Q^6$ is a better fit than the expected fit. However, there needs to be further research to determine the reason for the two data points that don't fall on the best fit line because it could change what the best fit is for this data set.
