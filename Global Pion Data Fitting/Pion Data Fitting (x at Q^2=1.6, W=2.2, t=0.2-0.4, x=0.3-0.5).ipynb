```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.320,0.381,0.396,0.426,0.476,0.311,0.323,0.318),(3.35,4.08,1.78,1.58,0.721,3.175,2.768,7.152),(0.41,0.41,0.18,0.19,0.060,0.361,0.272,0.356)])

x = pts[0,:]
y = pts[1,:]
yerror = pts[2,:]

m,b = np.polyfit(x,y,1)
print(m,b)

correlation_matrix = np.corrcoef(x,y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.plot(x, m*x + b, label='Best Fit Line (y = -21.997x + 11.190)($R^2$=0.462)')
plt.legend()
plt.errorbar(x,y,yerror,fmt=' ')
plt.title('x Dependence at $Q^2$=1.6, W=2.2, t=0.2-0.4, x=0.3-0.5', fontsize=20)
plt.xlabel('x', fontsize=15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize=15)
plt.show()
```

```python
x_2 = pts[0,:]**5
y_2 = np.log(pts[1,:])
y_2error = y_2*(pts[2,:]/pts[1,:])

m,b = np.polyfit(x_2,y_2,1)
print(m,b)

correlation_matrix = np.corrcoef(x_2,y_2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_2,y_2)
plt.plot(x_2, m*x_2 + b, label='Best Fit Line (y = -80.491x + 1.630)($R^2$=0.762)')
plt.legend()
plt.errorbar(x_2,y_2,y_2error,fmt=' ')
plt.title('$x^5$ Dependence at $Q^2$=1.6, W=2.2, t=0.2-0.4, x=0.3-0.5', fontsize=20)
plt.xlabel('$x^5$', fontsize=15)
plt.ylabel(r'log$\sigma$ (log$\frac{nbar}{c}$)', fontsize=15)
plt.show()
```
