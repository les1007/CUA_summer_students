```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.70,0.646,0.781,0.945,1.416,1.487),(10.4,17.51,15.79,13.63,6.44,4.179),(1.8,1.86,1.93,1.29,0.68,0.169)])

x_1 = pts[0,:]
y_1 = pts[1,:]
y_1error = pts[2,:]
print(x_1,y_1,y_1error)

m,b = np.polyfit(x_1,y_1,1)
plt.scatter(x_1,y_1)
plt.plot(x_1, m*x_1 + b, label='Best Fit Line (y = -12.609x + 23.881)($R^2$ = 0.7721)')
plt.legend()
plt.errorbar(x_1,y_1,y_1error,fmt=' ', color='blue')
plt.title('$Q^2$ Dependence at W=2.2, t=0.08', fontsize=20)
plt.xlabel('$Q^2$ ($GeV^2$)', fontsize=13)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize=13)
plt.show()

print(m,b)

correlation_matrix = np.corrcoef(x_1,y_1)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```
