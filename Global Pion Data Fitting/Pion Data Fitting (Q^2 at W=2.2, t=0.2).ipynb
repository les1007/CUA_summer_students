```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.70,1.667,2.279,1.658,1.610),(4.79,3.91,2.34,2.521,5.443),(0.98,0.44,0.21,0.110,0.468)])

x_1 = pts[0,:]
y_1 = pts[1,:]
y_1error = pts[2,:]
print(x_1)
print(y_1)
print(y_1error)

plt.figure(figsize=(8,5))
m,b = np.polyfit(x_1,y_1,1)
plt.scatter(x_1,y_1)
plt.plot(x_1, m*x_1 + b, label='Best Fit Line (y = -1.51x + 6.20)($R^2$=0.392)')
plt.errorbar(x_1,y_1,y_1error, fmt=' ', color='blue')
plt.legend()
plt.title('$Q^2$ Dependence at W=2.2, t=0.2', fontsize=20)
plt.xlabel('$Q^2$ ($GeV^2$)', fontsize=13)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize=13)
plt.show()

print(m,b)

correlation_matrix = np.corrcoef(x_1,y_1)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
x_max = (0.70**3,2.279**3)
x_max = np.array(x_max)
y_max = (pts[1,0]-pts[2,0],pts[1,2]+pts[2,2])
y_max = np.array(y_max)

x_min = (0.70**3,2.279**3)
x_min = np.array(x_min)
y_min = (pts[1,0]+pts[2,0],pts[1,2]-pts[2,2])
y_min = np.array(y_min)

x_2 = pts[0,:]**3
y_2 = pts[1,:]

plt.figure(figsize=(8,5))
m,b = np.polyfit(x_2,y_2,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)
plt.scatter(x_2,y_2)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_2, m*x_2 + b, label='Best Fit Line (y = -0.22x + 4.94)($R^2$=0.460)')
plt.plot(x_min, m1*x_min + b1, linestyle='--', dashes=(3,5), label='Min Slope (y = -0.32x + 5.88)')
plt.plot(x_max, m2*x_max + b2, linestyle='--', dashes=(3,5), label='Max Slope (y = -0.11x + 3.85)')
plt.errorbar(x_2,y_2,y_1error, fmt=' ', color='blue')
plt.legend()
plt.title('$(Q^2)^3$ Dependence at W=2.2, t=0.2', fontsize=20)
plt.xlabel('$(Q^2)^3$ (($GeV^2)^3$)', fontsize=13)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)', fontsize=13)
plt.show()

print(m,b)
print(m1,b1)
print(m2,b2)

correlation_matrix = np.corrcoef(x_2,y_2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
x_3 = pts[0,:]**3
y_3 = np.log(pts[1,:])
y_3error = y_3*(pts[2,:]/pts[1,:])

x_min = (x_3[0],x_3[2])
x_min = np.array(x_min)
y_min = (y_3[0]+y_3error[0], y_3[2]-y_3error[2])
y_min = np.array(y_min)
x_max = (x_3[0],x_3[2])
x_max = np.array(x_max)
y_max = (y_3[0]-y_3error[0], y_3[2]+y_3error[2])
y_max = np.array(y_max)

m,b = np.polyfit(x_3,y_3,1)
print(m,b)
m1,b1 = np.polyfit(x_min,y_min,1)
print(m1,b1)
m2,b2 = np.polyfit(x_max,y_max,1)
print(m2,b2)

correlation_matrix = np.corrcoef(x_3,y_3)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_3,y_3)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_3, m*x_3 + b, label='Best Fit Line (y = -0.06x + 1.61)($R^2$=0.508)')
plt.plot(x_min, m1*x_min + b1, linestyle='--', dashes=(5,3), label='Minimum Slope (y = -0.10x + 1.92)')
plt.plot(x_max, m2*x_max + b2, linestyle='--', dashes=(3,5), label='Maximum Slope (y = -0.03x + 1.26)')
plt.legend()
plt.errorbar(x_3,y_3,y_3error,fmt=' ', color='blue')
plt.title('$(Q^2)^3$ Dependence at W=2.2, t-0.2', fontsize =20)
plt.xlabel('$(Q^2)^3$ ($(GeV^2)^3$)', fontsize = 15)
plt.ylabel(r'ln$\sigma$ (ln$\frac{nbar}{c}$)')
plt.show()
```
