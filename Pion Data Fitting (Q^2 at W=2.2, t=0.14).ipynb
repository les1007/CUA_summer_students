```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.70,1.939,2.125,1.35,1.455,1.593,1.617),(5.81,3.175,2.768,4.62,7.152,4.292,2.807),(0.9,0.361,0.272,0.21,0.356,0.473,0.124)])
x_1 = pts[0,:]
y_1 = pts[1,:]
y_1error = pts[2,:]
            
print(x_1)
print(y_1)
print(y_1error)

x_min = (pts[0,0],pts[0,2])
x_min = np.array(x_min)
y_min = (pts[1,0]+pts[2,0],pts[1,2]-pts[2,2])
y_min = np.array(y_min)
x_max = (pts[0,0],pts[0,2])
x_max = np.array(x_max)
y_max = (pts[1,0]-pts[2,0],pts[1,2]+pts[2,2])
y_max = np.array(y_max)
```

```python
plt.figure(figsize=(8,5))
m,b = np.polyfit(x_1,y_1,1)
m1,b1 = np.polyfit(x_min,y_min,1)
m2,b2 = np.polyfit(x_max,y_max,1)

plt.scatter(x_1,y_1)
plt.scatter(x_min,y_min)
plt.scatter(x_max,y_max)
plt.plot(x_1, m*x_1 + b, label='Best Fit (y = -2.41 + 8.08) ($R^2$=0.447)')
plt.plot(x_min, m1*x_min + b1, dashes=(3,5), label='Min Slope (y = -2.96x + 8.78)')
plt.plot(x_max, m2*x_max + b2, dashes=(3,5), label='Max Slope (y = -1.31x + 5.83)')
plt.legend()
plt.errorbar(x_1,y_1,y_1error, fmt=' ', color='blue')
plt.title('$Q^2$ Dependence at W=2.2, t=0.14', fontsize=20)
plt.xlabel('$Q^2$ ($GeV^2$)', fontsize=15)
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c}$)',fontsize=15)
plt.show()

print(m,b)
print(m1,b1) 
print(m2,b2)

correlation_matrix = np.corrcoef(x_1,y_1)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```
